##  Membangun Model untuk Memprediksi Churn Pengguna
- Lihat prediksi apakah tingkat churn menggunakan model machine learning.


### Membagi Data Menjadi Train Set dan Validation Set Menggunakan Fungsi `train_test_split()`.󠀲󠀡󠀥󠀥󠀧󠀨󠀣󠀨󠀳
Di sini kita akan membagi beberapa yaitu:

- Memilih fitur-fitur untuk model dan memisahkan data menjadi set pelatihan dan pengujian.
- Standarisasi fitur menggunakan StandardScaler untuk memastikan fitur memiliki skala yang seragam.
- Membuat model Logistic Regression dan melatihnya menggunakan set pelatihan.
- Melakukan prediksi pada set pengujian.
- Mengevaluasi model menggunakan akurasi, confusion matrix, dan classification report.

```python
# Memilih fitur-fitur untuk model
features = churn_df.drop(['Churn'], axis=1)
target = churn_df['Churn']

# Memisahkan data menjadi set pelatihan (train) dan pengujian (test)
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Standarisasi fitur
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Membuat model Logistic Regression
model = LogisticRegression(random_state=42)
model.fit(X_train_scaled, y_train)

# Melakukan prediksi pada set pengujian
predictions = model.predict(X_test_scaled)

# Evaluasi model
accuracy = accuracy_score(y_test, predictions)
conf_matrix = confusion_matrix(y_test, predictions)
classification_rep = classification_report(y_test, predictions)

# Menampilkan hasil evaluasi
print(f'Accuracy: {accuracy:.2f}\n')
print('Confusion Matrix:')
print(conf_matrix)
print('\nClassification Report:')
print(classification_rep)
```

<font color = green> Kesimpulan Sementara </font>

1. Akurasi (Accuracy):
    - Akurasi model sebesar 92% menunjukkan bahwa model memiliki kemampuan yang baik dalam memprediksi kelas pelanggan yang churn dan tidak churn pada data pengujian.

2. Confusion Matrix:
    - True Positives (TP): 158 pelanggan yang diprediksi benar sebagai churn.
    - True Negatives (TN): 575 pelanggan yang diprediksi benar sebagai tidak churn.
    - False Positives (FP): 23 pelanggan yang sebenarnya tidak churn, tetapi diprediksi sebagai churn.
    - False Negatives (FN): 44 pelanggan yang sebenarnya churn, tetapi diprediksi sebagai tidak churn.

3. Precision, Recall, dan F1-Score:
    - Precision (presisi) untuk kelas 1 (churn) sebesar 0.87 menunjukkan seberapa baik model dapat memprediksi pelanggan yang sebenarnya churn dari yang diprediksi sebagai churn.
    - Recall (sensitivitas) untuk kelas 1 sebesar 0.78 menunjukkan seberapa baik model dapat menangkap pelanggan yang sebenarnya churn dari keseluruhan pelanggan yang sebenarnya churn.
    - F1-score sebesar 0.83 adalah harmonic mean dari precision dan recall, memberikan gambaran holistik tentang kinerja model.

4. Macro dan Weighted Average:
    - Macro avg dan weighted avg untuk precision, recall, dan f1-score memberikan nilai rata-rata yang dihitung dari masing-masing kelas (0 dan 1). Nilai ini memberikan gambaran kinerja model secara keseluruhan.

```python
# Menghitung nilai probabilitas positif (churn) dari model pada set pengujian
probs = model.predict_proba(X_test_scaled)[:, 1]

# Menghitung nilai False Positive Rate (FPR) dan True Positive Rate (TPR)
fpr, tpr, thresholds = roc_curve(y_test, probs)

# Menghitung Area Under the Curve (AUC)
roc_auc = auc(fpr, tpr)

# Membuat kurva ROC
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (AUC = {:.2f})'.format(roc_auc))
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.show()
```

<font color = green> Kesimpulan Sementara </font>

**Untuk Gambar sub bab 1.6.1**
- Secara umum, semakin mendekati sudut kiri atas, semakin baik modelnya. Jika AUC tinggi, itu menunjukkan bahwa model memiliki kemampuan membedakan yang baik antara kelas positif dan negatif.
- Kurva ROC ideal akan mendekati sudut kiri atas dari plot, yang mencerminkan nilai True Positive Rate (TPR) yang tinggi dan False Positive Rate (FPR) yang rendah. Semakin dekat kurva dengan sudut tersebut, semakin baik kinerja model.

### Latih Model Pada Train Set Dengan Dua Metode

```python
# Membuat model Regresi Logistik
model_logistic_regression = LogisticRegression(random_state=42)
model_logistic_regression.fit(X_train_scaled, y_train)

# Melakukan prediksi pada set pengujian
predictions_logistic_regression = model_logistic_regression.predict(X_test_scaled)

# Evaluasi model Regresi Logistik
accuracy_logistic_regression = accuracy_score(y_test, predictions_logistic_regression)
conf_matrix_logistic_regression = confusion_matrix(y_test, predictions_logistic_regression)
classification_rep_logistic_regression = classification_report(y_test, predictions_logistic_regression)

# Menampilkan hasil evaluasi Regresi Logistik
print(f'Accuracy Logistic Regression: {accuracy_logistic_regression:.2f}\n')
print('Confusion Matrix Logistic Regression:')
print(conf_matrix_logistic_regression)
print('\nClassification Report Logistic Regression:')
print(classification_rep_logistic_regression)
```

## Membuat Klaster Pengguna
- Kita akan pilih kolom yang berisi data churn dan tentukan klaster objeknya (pengguna).

```python
# Memilih kolom-kolom yang akan di-standardisasi
columns_to_standardize = ['Age', 'Lifetime', 'Avg_class_frequency_total', 'Avg_class_frequency_current_month', 'Avg_additional_charges_total']

# Memilih kolom yang berisi data churn
churn_column = 'Churn'

# Memisahkan data menjadi dua bagian: data yang akan di-standardisasi dan kolom Churn
cluster_data_to_standardize = churn_df[columns_to_standardize]
churn_labels = churn_df[churn_column]

# Inisialisasi objek StandardScaler
scaler = StandardScaler()

# Standardisasi data menggunakan fit_transform
cluster_data_standardized = scaler.fit_transform(cluster_data_to_standardize)

# Membuat DataFrame baru dengan data yang sudah di-standardisasi dan menambahkan kolom Churn
cluster_data_standardized_df = pd.DataFrame(cluster_data_standardized, columns=columns_to_standardize)
cluster_data_standardized_df[churn_column] = churn_labels
```

- **Avg_class_frequency_total (Frekuensi Rata-rata Kunjungan Total)**: Menunjukkan seberapa sering pelanggan mengunjungi pusat kebugaran dalam rentang waktu tertentu. Frekuensi kunjungan dapat mencerminkan tingkat keterlibatan dan kepuasan pelanggan terhadap layanan.


- **Avg_class_frequency_current_month (Frekuensi Rata-rata Kunjungan Bulan Ini)**: Merupakan indikator frekuensi kunjungan dalam satu bulan terakhir. Perubahan dalam frekuensi kunjungan mungkin menjadi sinyal potensial untuk perubahan perilaku pelanggan.


- **Avg_additional_charges_total (Total Biaya Tambahan Rata-rata)**: Menunjukkan seberapa banyak pelanggan mengeluarkan uang untuk layanan tambahan di pusat kebugaran, seperti kafe, barang atletik, kosmetik, atau pijat.
